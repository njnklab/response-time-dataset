```python
from data import Data
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
phq9 = Data('phq9')
gad7 = Data('gad7')
isi = Data('isi')
pss = Data('pss')
```

```python
def show_pca(data, ax):
    pca = PCA(n_components=2)
    df = data.df
    cols = [col for col in df.columns if col.startswith('time')]
    df_pca = pd.DataFrame(pca.fit_transform(df[cols]), columns=['pca1', 'pca2'])
    df_pca['score'] = df['score']
    sns.scatterplot(data=df_pca, x='pca1', y='pca2', hue='score', ax=ax)
    ax.set_title(data.scale_name)
    ax.set_xlabel('pca1')
    ax.set_ylabel('pca2')
    ax.set_xscale('log')
    ax.set_yscale('log')

fig, axes = plt.subplots(2, 2, figsize=(15, 15))
show_pca(phq9, axes[0, 0])
show_pca(gad7, axes[0, 1])
show_pca(isi, axes[1, 0])
show_pca(pss, axes[1, 1])
```
